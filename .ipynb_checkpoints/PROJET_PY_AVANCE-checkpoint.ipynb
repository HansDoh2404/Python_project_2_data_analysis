```python
import pandas as pd
```

```python
for col in data.columns :
   print(data[col].unique())
```

```python
# Dictionnaire de traduction
translation_dict = {
    'Location': 'Localisation',
    'Urban': 'Urbain',
    'Rural': 'Rural',
    'State': 'Ville',
    'Region': 'Région',
    'East': 'Est',
    'Northeast': 'Nord-Est',
    'Centre': 'Centre',
    'Midwest': 'Centre-Ouest',
    'Investment': 'Somme investie',
    'Construction': 'Type de construction',
    'Frame': 'Charpenterie',
    'Fire Resist': 'Résistant au feu',
    'Masonry': 'Maçonnerie',
    'Metal Clad': 'Construction en métal',
    'BusinessType': 'Type d\'entreprise',
    'Retail': 'Vente au détail',
    'Apartment': 'Appartement',
    'Farming': 'Ferme',
    'Hospitality': 'Hôtellerie',
    'Office Bldg': 'Bâtiment de bureaux',
    'Other': 'Autre',
    'Medical': 'Hôpitaux',
    'Organization': 'Société / Firme',
    'Construction': 'Architecture',
    'Service': 'Service Publique',
    'Education': 'Ecole',
    'Recreation': 'Restaurant',
    'Manufacturing': 'Industrie',
    'Earthquake': 'Séisme',
    'Yes': 'Oui',
    'No': 'Non',
    'Flood': 'Inondation',
    'Rating': 'Rentabilité'
}

# Traduire les colonnes
data.rename(columns=translation_dict, inplace=True)

# Traduire les valeurs
for column in data.columns:
    data[column] = data[column].map(translation_dict).fillna(data[column])
```

```python
data.to_excel("database.xlsx", index = False)
```
